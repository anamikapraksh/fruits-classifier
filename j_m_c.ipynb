```python
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv2D, MaxPooling2D
```

```python
from keras.utils import to_categorical
from sklearn.model_selection import train_test_split
```

```python
img_rows, img_cols = 28, 28
input_shape = (img_rows, img_cols, 1)
```

```python
cnn = Sequential()
cnn.add(Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=input_shape))
cnn.add(MaxPooling2D(pool_size=(2, 2)))
cnn.add(Dropout(0.2))
```

```python
cnn.add(Flatten())

cnn.add(Dense(128, activation='relu'))
cnn.add(Dense(2, activation='softmax'))
```

```python
cnn.compile(loss=keras.losses.categorical_crossentropy,
              optimizer=keras.optimizers.Adam(),
              metrics=['accuracy'])
```

```python
import tensorflow as tf
from keras.preprocessing.image import ImageDataGenerator
datagen_train = ImageDataGenerator()
datagen_validation = ImageDataGenerator()
```

```python
train_generator = datagen_train.flow_from_directory("/content/data/train",
                                                    target_size=(img_rows, img_cols),
                                                    
                                                    batch_size=30,
                                                    class_mode='categorical',
                                                    shuffle=True)

validation_generator = datagen_validation.flow_from_directory("/content/data/test",
                                                    target_size=(img_rows, img_cols),
                                                    
                                                    batch_size=30,
                                                    class_mode='categorical',
                                                    shuffle=False)
```

```python
from tensorflow.keras.models import Sequential
model = Sequential()
```

```python
model.add(tf.keras.layers.experimental.preprocessing.Rescaling(1./255, input_shape=(img_rows, img_cols,3)))
```

```python
model.add(tf.keras.layers.Conv2D(16,3,padding="same",activation="relu"))
model.add(tf.keras.layers.MaxPooling2D())
model.add(tf.keras.layers.Conv2D(32,3,padding="same",activation="relu"))
model.add(tf.keras.layers.MaxPooling2D())
model.add(tf.keras.layers.Conv2D(64,3,padding="same",activation="relu"))
model.add(tf.keras.layers.MaxPooling2D())
model.add(tf.keras.layers.Flatten())
```

```python
model.add(tf.keras.layers.LeakyReLU(alpha=0.5))
model.add(tf.keras.layers.Dropout(0.1))
model.add(tf.keras.layers.Dense(64,activation="relu",kernel_regularizer=tf.keras.regularizers.L1(0.01)))
model.add(tf.keras.layers.Dropout(0.1))
model.add(tf.keras.layers.Dense(2,activation="softmax"))
```

```python
import tensorflow
opt=tensorflow.keras.optimizers.Adam(learning_rate=0.05)
model.compile(optimizer="adam",loss=tf.keras.losses.CategoricalCrossentropy(),metrics=['accuracy'])
model.summary()
```

```python
history = model.fit_generator(generator=train_generator,
                                
                                epochs=50,
                                # validation_data = validation_generator,
                                )
```

```python
model.evaluate(validation_generator)
```

```python
model.save("model.h5")
```
